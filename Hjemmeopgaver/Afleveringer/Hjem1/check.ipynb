```python
import numpy as np
import matplotlib.pyplot as plt

# Definerer funktionen
def f(x):
    return 4*x**2 + 4*np.abs(x - 1)

# Definer et interval for x-værdierne
x = np.linspace(-2, 3, num=500)

# Beregn y-værdierne for funktionen
y = f(x)

# Plot funktionen
plt.plot(x, y, label='$f(x) = 4x^2 + 4|x - 1|$')
plt.title('Plot af funktionen $f(x) = 4x^2 + 4|x - 1|$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Definer komplekse tal
z0 = 1.1714455936572892 + 0.20476519107020205j
z1 = -0.763054654106214 + 0.9121190477234544j
z2 = -0.40839093955107547 - 1.1168842387936562j
z3 = 1.1714455936572892 + 0.2047651910702019j

# Beregn deres positions i enhedscirklen
positions = [z0, z1, z2, z3]

# Plot punkterne
plt.figure(figsize=(8, 8))
for pos in positions:
    plt.scatter(pos.real, pos.imag, color='blue')

# Tegn enhedscirklen
circle = plt.Circle((0, 0), radius=1, color='red', fill=False)
plt.gca().add_artist(circle)

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0 , color='black',linewidth=0.5)
plt.title('Plot af komplekse tal i enhedscirklen')
plt.xlabel('Realdel')
plt.ylabel('Imaginærdel')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.show()
```
