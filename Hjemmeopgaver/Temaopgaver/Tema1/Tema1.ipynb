Lad os starte med at importere de nødvendige biblioteker og definere funktionen:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definerer funktionen f(z) = i * z
def f(z):
    return 1j * z
```

Nu indtegner vi de givne komplekse tal (2, 1 + i og -2 + 3i) i den komplekse talplan:

```python
# De givne komplekse tal
tal = [2, 1 + 1j, -2 + 3j]

# Plot tallene
plt.scatter([z.real for z in tal], [z.imag for z in tal], color='blue')
for z in tal:
    plt.text(z.real, z.imag, str(z), fontsize=12, ha='right', va='bottom')

# Tegn enhedscirklen (radius 1)
circle = plt.Circle((0, 0), radius=1, color='gray', fill=False)
plt.gca().add_artist(circle)

# Indstil akserne og vis figuren
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.grid(True)
plt.title('Komplekse tal i den komplekse talplan')
plt.xlabel('Realdel')
plt.ylabel('Imaginærdel')
plt.show()
```

Derefter beregner vi funktionsværdierne for hvert af disse tal:

```python
# Beregn f(2), f(1 + i) og f(-2 + 3i)
resultater = [f(z) for z in tal]

# Udskriv resultaterne
for z, resultat in zip(tal, resultater):
    print(f"f({z}) = {resultat}")
```

### b

 Vi skal bruge, at $ z = |z|e^{i\text{arg}(z)} $, og vi vil vise, at $ f(z) = |z|e^{i(\text{arg}(z) + \frac{\pi}{2})} $. 

For det første skal vi huske på, hvordan man kan repræsentere en kompleks størrelse i polære koordinater. Enhver komplekst tal $ z $ kan skrives som $ z = |z|e^{i\theta} $, hvor $ |z| $ er tallets modulus (eller absolut værdi) og $ \theta $ er dets argument (eller fasevinkel).

### Trigonometriske identiteter
Vi starter med at bruge Euler's formel, som siger:
$ e^{i\theta} = \cos(\theta) + i\sin(\theta) $

### Rotation af komplekse tal
Lad os nu betragte funktionen $ f(z) $. Vi vil vise, at $ f(z) $ er en rotation af $ z $ med en vinkel på $ \frac{\pi}{2} $ mod uret. 

For at gøre dette, lad os kigge på argumentet (eller fasevinklen) til $ z $. Hvis $ z = |z|e^{i\text{arg}(z)} $, så er argumentet af $ z $ givet ved $ \text{arg}(z) $.

### Rotation i det komplekse plan
Når vi roterer et punkt med en vinkel $ \theta $ mod uret, skifter argumentet (eller fasevinklen) fra $ \phi $ til $ \phi + \theta $. 

I vores tilfælde roterer vi $ z $ med en vinkel på $ \frac{\pi}{2} $ mod uret. Det betyder, at det nye argument bliver:
$ \text{arg}(z) + \frac{\pi}{2} $

### Udregning af $ f(z) $
Nu kan vi skrive $ f(z) $ som:
$ f(z) = |z|e^{i(\text{arg}(z) + \frac{\pi}{2})} $

Dette er en rotation af det komplekse tal $ z $ med en vinkel på $ \frac{\pi}{2} $ mod uret, idet vi tilføjer $ \frac{\pi}{2} $ til argumentet.

### Konklusion
Det viser, at hvis man roterer et komplekst tal $ z $ med en vinkel på $ \frac{\pi}{2} $ mod uret i det komplekse plan, så bliver resultatet givet ved:
$ f(z) = |z|e^{i(\text{arg}(z) + \frac{\pi}{2})} $

Dette er en rent geometrisk og algebraisk argument for den roterende effekt af $ f(z) $ på et komplekst tal.

### c

 Vi definerer funktionen $ f_\alpha : \mathbb{C} \to \mathbb{C} $ ved $ z \mapsto e^{i\alpha} \cdot z $, hvor $ \alpha \in \mathbb{R} $.

### Trin 1: Udtryk for $ f_\alpha(z) $
Vi skal vise, at $ f_\alpha(z) = |z|e^{i(\text{arg}(z) + \alpha)} $.

For et vilkårligt kompleks tal $ z = |z|e^{i\text{arg}(z}) $, gælder der:
$ f_\alpha(z) = e^{i\alpha} \cdot |z|e^{i\text{arg}(z)} $

### Trin 2: Udregning af $ f_\alpha(z) $
Vi udnytter, at produktet af to komplekse tal er et kompleks tal med modulus lig produktet af deres modulier og argument lig summen af deres argumenter. Derfor har vi:
$ e^{i\alpha} \cdot |z|e^{i\text{arg}(z)} = |z|e^{i(\text{arg}(z) + \alpha)} $

### Trin 3: Rotation i det komplekse plan
Nu ved vi, at $ f_\alpha(z) = |z|e^{i(\text{arg}(z) + \alpha)} $. Dette udtryk beskriver et punkt på enhedscirklen med radius $ |z| $ og argument $ \text{arg}(z) + \alpha $.

### Trin 4: Rotationens effekt
En rotation i det komplekse plan med vinklen $ \alpha $ mod uret kan beskrives som en multiplikation af det komplekse tal med $ e^{i\alpha} $. Derfor er $ f_\alpha(z) $ en rotation af $ z $ med en vinkel på $ \alpha $ mod uret.

### Trin 5: Konklusion
Vi har vist, at $ f_\alpha(z) = |z|e^{i(\text{arg}(z) + \alpha)} $, hvilket betyder, at funktionen $ f_\alpha $ virker som en rotation med vinklen $ \alpha $ mod uret.

Derfor er konklusionen, at måden, hvorpå funktionen $ f_\alpha $ virker på den komplekse talplan, netop er en sådan rotation.

 For at løse dette problem, skal vi finde værdien af α, således at $ f_{\alpha}(\sqrt{2}/2 + i\sqrt{2}/2) = 1/2 + i\sqrt{3}/2 $. Vi kan starte med at beregne argumenterne for de to komplekse tal.

1. Beregn argumentet for $ \sqrt{2}/2 + i\sqrt{2}/2 $:
   - Argumentet (eller fasen) af et komplekst tal $ a + bi $ er givet ved $ \theta = \tan^{-1}(b/a) $.
   - For $ \sqrt{2}/2 + i\sqrt{2}/2 $, har vi $ a = \sqrt{2}/2 $ og $ b = \sqrt{2}/2 $.
   - Derfor er argumentet: 
     $
     \theta_1 = \tan^{-1}\left(\frac{\sqrt{2}/2}{\sqrt{2}/2}\right) = \tan^{-1}(1) = \frac{\pi}{4}
     $

2. Beregn argumentet for $ 1/2 + i\sqrt{3}/2 $:
   - For $ 1/2 + i\sqrt{3}/2 $, har vi $ a = 1/2 $ og $ b = \sqrt{3}/2 $.
   - Derfor er argumentet:
     $
     \theta_2 = \tan^{-1}\left(\frac{\sqrt{3}/2}{1/2}\right) = \tan^{-1}(\sqrt{3}) = \frac{\pi}{3}
     $

Nu har vi de to argumenter $ \theta_1 = \frac{\pi}{4} $ og $ \theta_2 = \frac{\pi}{3} $.

3. Beregn forskellen i argumenter:
   $
   \Delta\theta = \theta_2 - \theta_1 = \frac{\pi}{3} - \frac{\pi}{4} = \frac{4\pi}{12} - \frac{3\pi}{12} = \frac{\pi}{12}
   $

4. Find α således at $ f_{\alpha}(\sqrt{2}/2 + i\sqrt{2}/2) = 1/2 + i\sqrt{3}/2 $:
   - Funktionen $ f_{\alpha}(z) $ roterer punktet z med en vinkel på $ \alpha $.
   - Vi skal have:
     $
     f_{\alpha}\left(\frac{\sqrt{2}}{2} + i\frac{\sqrt{2}}{2}\right) = 1/2 + i\sqrt{3}/2
     $
   - Argumentet for $ \frac{\sqrt{2}}{2} + i\frac{\sqrt{2}}{2} $ er $ \frac{\pi}{4} $, og argumentet for $ 1/2 + i\sqrt{3}/2 $ er $ \frac{\pi}{3} $.
   - Forskellen i argumenter er:
     $
     \Delta\theta = \frac{\pi}{3} - \frac{\pi}{4} = \frac{\pi}{12}
     $
   - Derfor skal $ \alpha $ være lig med forskellen, dvs. $ \alpha = \frac{\pi}{12} $.

Men vi er også givet en ligning: 
$$
\sqrt{3}/2 + i/2 = e^{i(\pi/6 + alpha)}
$$
Løs for α:
$$
e^{i(\pi/6 + \alpha)} = \frac{\sqrt{3}}{2} + i\frac{1}{2}
$$
Denne ligning kan omskrives til:
$$
\cos(\pi/6 + \alpha) = \frac{\sqrt{3}}{2}
$$
$$
\sin(\pi/6 + \alpha) = \frac{1}{2}
$$
Så derfor må:
$$
\pi/6 + \alpha = \pi/6 + 5\pi/6 = \pi
$$
Altså, $ \alpha = 0 $ og så er løsningen i intervallet [0;2π]:
$$
\alpha = -\frac{5}{6}\pi
$$

Men dette ville ikke give mening hvis opgaven var at finde en vinkel, der kunne tilføjes til $\pi/6$ for at få argumentet for den anden komplekse tal. I stedet skulle vi bruge trigonometriske identiteter og ligninger til at isolere $\alpha$.

For at finde $\alpha$, kan vi se på cosinus og sinus af de to sider af lighedstegnet:
$$
\cos(\pi/6 + \alpha) = \frac{\sqrt{3}}{2}
$$
$$
\sin(\pi/6 + \alpha) = \frac{1}{2}
$$
Vi ved, at $\cos(\pi/6) = \frac{\sqrt{3}}{2}$ og $\sin(\pi/6) = \frac{1}{2}$. Så derfor må:
$$
\cos(\pi/6 + \alpha) = \cos(\pi/6) \cdot \cos(\alpha) - \sin(\pi/6) \cdot \sin(\alpha) = \frac{\sqrt{3}}{2}
$$
$$
\sin(\pi/6 + \alpha) = \sin(\pi/6) \cdot \cos(\alpha) + \cos(\pi/6) \cdot \sin(\alpha) = \frac{1}{2}
$$
Vi ved, at $\cos(\pi/6) = \frac{\sqrt{3}}{2}$ og $\sin(\pi/6) = \frac{1}{2}$, så:
$$
\frac{\sqrt{3}}{2} \cdot \cos(\alpha) - \frac{1}{2} \cdot \sin(\alpha) = \frac{\sqrt{3}}{2}
$$
$$
\frac{1}{2} \cdot \cos(\alpha) + \frac{\sqrt{3}}{2} \cdot \sin(\alpha) = \frac{1}{2}
$$
Vi kan løse disse to ligninger sammen med hinanden for at finde $\cos(\alpha)$ og $\sin(\alpha)$. Men da vi allerede ved, at $\cos(\pi/6 + \alpha) = \frac{\sqrt{3}}{2}$ og $\sin(\pi/6 + \alpha) = \frac{1}{2}$, kan vi konkludere, at:
$$
\alpha = 0
$$
Derfor er løsningen i intervallet [0;2π]:
$$
\alpha = -\frac{5}{6}\pi
$$

### e

 For at vise, at der ikke eksisterer et $\alpha \in \mathbb{R}$ således at $f_\alpha(1+i) = 2+3i$, kan vi starte med at beregne modulus af de to komplekse tal.

1. Beregn modulus af $1 + i$:
   $
   |1 + i| = \sqrt{1^2 + 1^2} = \sqrt{2}
   $

2. Beregn modulus af $2 + 3i$:
   $
   |2 + 3i| = \sqrt{2^2 + 3^2} = \sqrt{4 + 9} = \sqrt{13}
   $

Nu skal vi finde $\alpha$ således at:
$
f_\alpha(1+i) = |1+i| e^{i(\arg(1+i) + \alpha)} = 2 + 3i
$

Modulusen af $1+i$ er $\sqrt{2}$, og vi skal have:
$
\sqrt{2} e^{i(\arg(1+i) + \alpha)} = 2 + 3i
$

Beregn argumentet for $1+i$:
$
\arg(1+i) = \tan^{-1}\left(\frac{1}{1}\right) = \frac{\pi}{4}
$

Så vi har:
$
\sqrt{2} e^{i(\frac{\pi}{4} + \alpha)} = 2 + 3i
$

Modulusen på venstre side er $\sqrt{2}$, og modulusen på højre side er $\sqrt{2+3^2} = \sqrt{13}$. For at de to sider kan være lig hinanden, skal modulusen på venstre side også være $\sqrt{13}$:
$
\sqrt{2} e^{i(\frac{\pi}{4} + \alpha)} = \sqrt{13}
$

Dette er umuligt, da $e^{i(\frac{\pi}{4} + \alpha)}$ kun kan have en værdi mellem $-1$ og $1$. Derfor kan der ikke findes et $\alpha$ som opfylder denne ligning.

Derfor konkluderer vi, at der ikke eksisterer noget $\alpha \in \mathbb{R}$ således at $f_\alpha(1+i) = 2+3i$.

Pointen med denne del er, at multiplikation af et komplekst tal med a ∈ C har en smuk
geometrisk fortolkning i den komplekse talplan: man skalerer med en faktor |a| og roterer
mod uret med en vinkel på arg(a).

### 1

 Funktionen $ h_b : \mathbb{C} \to \mathbb{C} $ defineret ved $ z \mapsto z + b $ har effekten en translation langs vektoren $ (\text{re}(b), \text{im}(b)) $.

### 2

 Vi starter med at skrive funktionerne $g_a$ og $h_b$ på deres eksplicitte former. Funktionen $g_a$ er defineret som $g_a(z) = a \cdot z$, hvor definitionsmængden er alle reelle tal, dvs., $\mathbb{R}$, mens dispositionsmængden også er $\mathbb{R}$. På samme måde er funktionen $h_b$ defineret som $h_b(y) = y + b$, hvor definitionsmængden er igen alle reelle tal, dvs., $\mathbb{R}$, mens dispositionsmængden også er $\mathbb{R}$.

Vi skal nu finde den sammensatte funktion $(h_b \circ g_a)(z)$. For at gøre dette bruger vi definitionen af sammensat funktion: 
$(h_b \circ g_a)(z) = h_b(g_a(z))$.

Vi indsætter $g_a(z)$ i $h_b$. Da $g_a(z) = a \cdot z$, får vi:
$h_b(a \cdot z)$.

Nu bruger vi definitionen af $h_b$ til at udregne værdien af denne funktion:
$h_b(a \cdot z) = a \cdot z + b$.

Derfor har vi fundet, at $(h_b \circ g_a)(z) = a \cdot z + b$.

### 3

 Svaret er: Funktionen $ f_{a,b}: \mathbb{C} \to \mathbb{C} $ defineret ved $ z \mapsto az + b $ har en geometrisk effekt på den komplekse talplan, der kan beskrives som følger:

1. **Skalering og Rotation:**
   - Først skaleres det givne komplekse tal $ z $ med en faktor svarende til den absolutte værdi af $ a $, dvs. $ |a| $. Dette betyder, at hvis $ a = re^{i\theta} $ (hvor $ r $ er modulus og $ \theta $ er argumentet eller fasevinklen), så skaleres tallet med faktoren $ r $.
   - Derefter roteres det komplekse tal mod uret med en vinkel på $ \text{arg}(a) $ eller $ \theta $. Rotationen foregår i den matematisk positive retning (mod uret), hvilket er den konventionelle definition af argumentet.

2. **Translation:**
   - Derefter translateres det resulterende punkt langs vektoren $ (\text{Re}(b), \text{Im}(b)) $. Dette betyder, at der til det roterede og skalerede tal lægges den reelle del af $ b $ (Re(b)) og den imaginære del af $ b $ (Im(b)), svarende til vektoren i det komplekse plan.

Dette betyder, at hvis du tager et komplekst tal $ z $, først roteres og skaleres efter en bestemt måde som beskrevet ovenfor (altså med hensyn til $ a $ og derefter $ b $), så vil resultatet være et nyt komplekst tal, der repræsenterer det oprindelige tal transformeret på denne specielle måde.
